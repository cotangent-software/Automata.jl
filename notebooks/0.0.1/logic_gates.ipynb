```julia
using Automata
using Plots
using Images
```

```julia
code = 2530
colors = 4
```

Below are the initial conditions for a NAND gate. These can be modified in a manner such that the period stays as multiples of two of the following gaps.

```julia
function plot_nand(op1, op2)
        return plot(
        Gray.(1 .- (Automata.to_array(totalistic(code, 30, colors, UInt8[2, 1, 0, 0, 0, 0,   (op1 ? UInt8[2, 1] : UInt8[0, 0])..., zeros(UInt8, 6)..., (op2 ? UInt8[2, 1] : UInt8[0, 0])..., zeros(UInt8, 5)..., 1, 2, zeros(UInt8, 3)..., 2, 2, 1, 1, 2])))[:, 30:60]),
        size=(200, 200),
        fmt = :png,
        title = "Â¬ ($(Int(op1)) & $(Int(op2))) = $(Int(!(op1 && op2)))"
    )
end
```

```julia
p1 = plot_nand(false, false)
p2 = plot_nand(false, true)
p3 = plot_nand(true, false)
p4 = plot_nand(true, true)
plot(p1, p2, p3, p4, layout=(2, 2), fmt=:png, size=(400, 400))
```
